MERGING OF THE DATASETS:

```python
import pandas as pd
# Load cleaned datasets
customers = pd.read_csv(r"C:\Users\Tanvi\Downloads\cleaned_customers.csv", encoding='ISO-8859-1')
sales = pd.read_csv(r"C:\Users\Tanvi\Downloads\cleaned_Sales1.csv")
stores = pd.read_csv(r"C:\Users\Tanvi\Downloads\cleaned_stores.csv")
products = pd.read_csv(r"C:\Users\Tanvi\Downloads\cleaned_product.csv")
exchange_rates = pd.read_csv(r"C:\Users\Tanvi\Downloads\cleaned_Exchange.csv")
```

```python
# Merging dataframes
merged_data = pd.merge(sales, customers, on="CustomerKey",how="inner")
merged_data = pd.merge(merged_data, products, on="ProductKey", how="inner")
merged_data = pd.merge(merged_data, stores,on=['StoreKey','Country','State'],how='inner')
```

```python
merged_data
```

```python
# Merging exchange rates with merged data
merged_data = pd.merge(merged_data, exchange_rates,left_on=['Order_Date','Currency_Code'],right_on=['Date','Currency'], how="inner")
```

```python
merged_data
```

```python
merged_data.to_csv(r'C:\Users\Tanvi\Downloads\merged_data1.csv', index=False)
```

```python
mycursor=mydb.cursor()
columns = merged_data.columns
column_types = []

# Define a mapping from pandas dtype to SQL type
type_mapping = {
     "int64": "INT",
     "float64": "FLOAT",
     "object": "VARCHAR(255)",
     "datetime64[ns]": "DATETIME",
  }

# Generate column type definitions
for col in columns:
    dtype = str(merged_data[col].dtype)
    sql_type = type_mapping.get(dtype, "VARCHAR(255)")
    if col == "Order_Numbe":  #Making order_no as the primary key column
        column_types.append(f"`{col}` {sql_type} PRIMARY KEY")
    else:
        column_types.append(f"`{col}` {sql_type}")

# Create SQL statement for table creation
table_query = f"""
   CREATE TABLE IF NOT EXISTS Electronic_table (
      {', '.join(column_types)}
  );
  """

# execute the SQL statement
mycursor.execute(table_query)
mydb.commit()

# Inserting  Merged_data in into Table query
query=f""" 
INSERT INTO Electronic_table ({', '.join([f'`{col}`' for col in columns])})
VALUES ({', '.join(['%s'] * len(columns))})
"""
# Insert the data
for index, row in merged_data .iterrows():
    mycursor.execute(query, tuple(row))

# Commit the transaction
mydb.commit()

# Close the cursor and connection
mycursor.close()
mydb.close()
print("dataset successfully stored in the MySQl database.")
```
